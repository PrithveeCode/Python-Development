### **1. REGEX Expressions**
- REGEX Expressions Are String Operations.
- Regular expressions are almost their own little programming language 
- Used for searching and parsing strings. 

```python
import re
# Import Regex Library To Get Started!.

'''
@brief: Search for lines that contain 'From'
        This Is Simple Example For REGEX.
'''
fd = open('../Samples/2.AvgData.txt', 'r')
for line in fd:
    line = line.rstrip()
    if re.search('From', line):
        print(line)
fd.close()

del(fd)
del(line)
```

#### **1.1 REGEX Additional Commands**

- The regular expressions efficiency comes when we add special characters to the search string 
  that allow us to more precisely control which lines match the string.
  
- Adding these special characters to our regular expression allow us to do sophisticated matching 
  and extraction while writing very little code.

```python
'''
@brief: Search For Lines That Startswith 'From'
@Note : We can Also use: .startswith() Function/Method provided by string Class.
'''
fd = open('../Samples/2.AvgData.txt')
for line in fd:
    line = line.rstrip()
    if (re.search('^From:', line)):
        # '^' Means -> The Start of The Line.
        print(line)
fd.close()

del(fd)
del(line)
```

```python
'''
@brief: Search For The Lines That StartsWith: 'F' followed by 2 Unknown Characters & Then Follwed By
        'm' Character.

@note : In the following example, the regular expression F..m: would match any of the
        strings “From:”, “Fxxm:”, “F12m:”, or “F!@m:”, etc.
'''

fd = open('../Samples/2.AvgData.txt', 'r')
for line in fd:
    line = line.rstrip()
    if (re.search('^F..m', line)):
        print(line)
```

```python
'''
@brief: These special characters mean that instead of matching a single
        character in the search string, 
        They Match Zero-Or-More Characters  :(In The Case Of The Asterisk(*)) Or,
        One-Or-More Of The Characters       :(In The Case Of The Plus Sign(+)).
'''

fd = open('../Samples/2.AvgData.txt', 'r')
for line in fd:
    line = line.rstrip()
    if (re.search('^From:.+@', line)): 
        print(line)

# This Will Search One or More Characters (.+) Followed By An @ Sign. 
# "From: stephen.marquard@uct.ac.za"

# You can think of the .+ wildcard as expanding to match all the characters between
# the colon character and the at-sign.

fd.close()
del(fd)
del(line)
```

```python
strData  = 'A message from csev@umich.edu to cwen@iupui.edu about meeting @2PM'
# \S means : Non-WhiteSpace Character!.
findRegex = re.findall('\S+@\S+', strData)
# This Means The REGEX Will look for -> A Word Which: 
#   1. Starts With "Non-WhiteSpace" Character(s). 
#   2. Contains A Character "@"
#   3. Ends With One or More "Non-WhiteSpace" Character(s).
print(findRegex)

# The regular expression would match twice (csev@umich.edu and cwen@iupui.edu),
# but it would not match the string “@2PM” because there are no "Non-Blank" characters before the at-sign.

del(findRegex)
del(strData)
```

```python
# Search For Lines -> That Have an Sign Between Character.
fd = open('../Samples/2.AvgData.txt')
for line in fd:
    line = line.rstrip()
    rout = re.findall('\S+@\S+', line)

    if (len(rout) > 0):
        print(rout)

fd.close()
del(rout)
del(line)
del(fd)
```

#### **1.2 Complex REGEX Queries!**
- In Above Code! -> We Can se Some Components such as ['<source@collab.sakaiproject.org>;'] 
- To Remove Those Un-Needed Characters!.

```python
# Here is Our Query, [a-zA-Z0-9]\S*@\S*[a-zA-Z] Kinda Advanced Right-?
# I believe [] Square Brackets! -> Are SubStrings Information.ss

'''
    @brief: This Query Means: The Line Should:
            1. Start With A Character -> That's A Lower or upperCase Alphabet or 0-9 Numbers.
            2. Followed By A One or More Non-WhiteSpace Character(s).
            3. Followed By A One Symbol "@".
            4. Followed By A Zero Or More Non - White Characters!.
'''
fd = open('../Samples/2.AvgData.txt', 'r')
for line in fd:
    line = line.rstrip()
    rout = re.findall('[a-zA-Z0-9]\S+@\S*[a-zA-Z]', line)
    if len(rout) > 0:
        print(rout)

'''
    --- @brief: Now, As We Can Notice -> 
    --- This Element Didn't Get Completely Removed in Below Example! -> ['<source@collab.sakaiproject.org>;']
    --- But, Speicifially it Picked The String B/w "<" & ">". (i.e. ['source@collab.sakaiproject.org'])

'''

fd.close()
del(line)
del(fd)
del(rout)
```

#### 1.3 Parenthesis {} in REGEX.

- When we add parenthesis to the REGEX statements, They become more useful.
- But, when we use -> findall() they are ignored!.
- In Simple Words Parenthesis is '^X\S*: (<This is What I Want/>)
- In definition: Parenthesis Indicate That Even If might be searching for a 
  whole string you nly want a part of it.

```python
fd = open('../Samples/2.AvgData.txt', 'r')
for line in fd:
    line = line.rstrip()
    rout = re.findall('^X\S+: ([0-9.]+)', line)
    if len(rout) > 0:
        print(rout)

fd.close()
del(line)
del(rout)


'''
^ :       Matches the beginning of a line
$ :       Matches the end of the line
. :       Matches any character
\s:       Matches whitespace
\S:       Matches any non-whitespace character
* :       Repeats a character zero or more times
*?:       Repeats a character zero or more times.  (non-greedy)
+ :       Repeats a character one or more times
+?:       Repeats a character one or more times.   (non-greedy)

[aeiou] : Matches a single character in the listed set
[^XYZ]  : Matches a single character not in the listed set
[a-z0-9]: The set of characters can include a range
( :       Indicates where string extraction is to start
) :       Indicates where string extraction is to end
'''
```

```python
str1 = "Hello My Name Is Prithvee."
rout = re.findall('[aeiou].+ ', str1)
print(rout)
```

#### **2. Assignment:**

- The basic outline of this problem is to read the file, look for integers using the re.findall(), looking for a regular expression of '[0-9]+' and then converting the extracted strings to integers and summing up the integers.



```python
sum = 0
def handle_data(listDat):
    global sum 
    # This Tells The Function To Use Global Function Instead of Local Variable.
    if len(listDat) > 0:
        for numStr in listDat:
            sum += int(numStr)
            

fd = open('../Samples/4.regex_sum_1841118.txt')
for line in fd:
    line = line.rstrip()
    rout = re.findall('[0-9]+', line)
    handle_data(rout)

print(f"Sum is {sum}.")

fd.close()
del(fd)
del(sum)
del(line)
del(rout)
```
